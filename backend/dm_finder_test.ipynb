# DM Finder - Surgical Query Tester
This notebook allows you to test the improved "Geographically Fenced" search queries for finding decision makers using Tavily.

```python
import os
from dotenv import load_dotenv
from tavily import TavilyClient
import json

# Load API Key from backend/.env
load_dotenv()
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")

if not TAVILY_API_KEY:
    print("❌ TAVILY_API_KEY not found in .env file!")
else:
    tavily = TavilyClient(api_key=TAVILY_API_KEY)
    print("✅ Tavily Client Initialized")
```

```python
def test_dm_queries(company_name, location):
    print(f"--- Testing for {company_name} in {location} ---\n")
    
    # Construction of Surgical Queries with Location Fencing
    queries = [
        f'site:linkedin.com/in "{company_name}" "{location}" ("CEO" OR "CTO" OR "Founder" OR "Managing Director")',
        f'site:linkedin.com/in "{company_name}" "{location}" ("VP" OR "Director" OR "Manager")'
    ]
    
    results = []
    
    for i, q in enumerate(queries):
        print(f"Executing Query {i+1}: {q}")
        # Using search_depth='advanced' (Tavily Advanced) as used in the production agent
        response = tavily.search(query=q, search_depth="advanced", max_results=5)
        
        print(f"Found {len(response['results'])} results.")
        for res in response['results']:
            print(f"- Title: {res['title']}")
            print(f"  URL: {res['url']}")
            print(f"  Snippet: {res['content'][:150]}...")
            print("")
        results.append(response)
        
    return results

# TEST HERE
company = "Scintilla Digital Academy"
city = "Hyderabad"

test_dm_queries(company, city)
```
