# Tavily API - Quick Start Guide
This notebook demonstrates how to use the Tavily Search API for high-quality, AI-optimized research.

```python
import os
from dotenv import load_dotenv
from tavily import TavilyClient
import json

# 1. Load your API Key
# This assumes your .env file is in the 'backend' folder
load_dotenv("backend/.env")
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")

if not TAVILY_API_KEY:
    print("❌ TAVILY_API_KEY not found!")
else:
    tavily = TavilyClient(api_key=TAVILY_API_KEY)
    print("✅ Tavily Client Initialized")
```

### 1. Basic Search
A simple search that returns titles, URLs, and snippets.

```python
query = 'GenAI99 executive leadership team'
response = tavily.search(query=query, max_results=5)

for result in response['results']:
    print(f"Title: {result['title']}")
    print(f"URL: {result['url']}")
    print(f"Snippet: {result['content'][:200]}...\n")
```

### 2. Advanced AI Search (Recommended for Agents)
Uses `search_depth='advanced'` and includes `include_raw_content` for better LLM processing.

```python
query = "Founder and leadership team of GenAI99 Hyderabad"
response = tavily.search(
    query=query, 
    search_depth="advanced", 
    max_results=5, 
    include_raw_content=True
)

print(f"Found {len(response['results'])} results with advanced search.\n")
print(json.dumps(response['results'][0], indent=2))
```

### 3. Domain Restricted Search (Solving Partial Results)
To solve the issue of 'partial results', you MUST use `include_raw_content=True`. This tells Tavily to scrape the full page content into a field called `raw_content` instead of just returning a snippet.

```python
query = 'www.gnanamai.com'
response = tavily.search(
    query=query, 
    search_depth="advanced",
    #include_domains="*", 
    include_raw_content=True, # <--- CRITICAL for full results
    max_results=5
)

for r in response['results']:
    print(f"--- RESULT: {r['title']} ---")
    print(f"URL: {r['url']}")
    print(f"CLEAN SNIPPET: {r['content']}\n")
    
    # ACCESS FULL DATA
    if 'raw_content' in r and r['raw_content']:
        # Printing first 2000 chars of the full scraped data
        print(f"FULL CONTENT PREVIEW: \n{r['raw_content'][:2000]}...\n")
    else:
         print("⚠️ No raw content found for this link.")
    
    print("=" * 80)
```
